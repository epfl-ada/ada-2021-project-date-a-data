```python
#plot only USA and then check for global ...

quotes_mapped='quotebank_politicians_USA_mapped.json.bz2'
WIKI_DATA_FILTERED='filtered_politician_labeled_us_v3.json.bz2' #catalogue containing all US politician

df_full=pd.read_json(quotes_mapped, lines=True)
catalogue_df=pd.read_json(WIKI_DATA_FILTERED, lines=True)
catalogue_df.head(3)
```

```python
df_full.head(3)
```

```python
df=df_full[['quoteID','qids','mentions_qids','speaker']] #extract info of quotations taking only USA politicians mentions of USA politician in the US catalogue
#df=df.loc[:500,:]
men_df=pd.DataFrame()
for i,row in df.iterrows():
    if row.qids[0] in catalogue_df.qid.values:
        for mention_qid in row.mentions_qids:
            if mention_qid in catalogue_df.qid.values:
                mention=pd.Series([row.quoteID,row.qids[0],mention_qid,row.speaker])
                men_df=men_df.append(mention,ignore_index=True)
                if i %10000==0: print(i)
men_df.rename(columns={0:'quoteID',1:'qid',2:'mention_qid',3:'speaker'},inplace=True)
print(men_df.head())
print(f'men_df shape {men_df.shape}')
```

```python
#plot for global ...

quotes_world='quotes_mentions_world_compact.json.bz2'
WIKI_DATA_FILTERED='filtered_politician_labeled_us_v3.json.bz2' #catalogue containing all US politician

df_world=pd.read_json(quotes_world, lines=True)
catalogue_df=pd.read_json(WIKI_DATA_FILTERED, lines=True)
catalogue_df.head(3)
```

```python
df1=df_world[['quoteID','qids','speaker','mentions_qids','mentions','country']] #extract info of quotations taking only USA politicians mentions of USA politician in the US catalogue
mentions_w_df=pd.DataFrame()
for i,row in df1.iterrows():
    for j,mention_qid in enumerate(row.mentions_qids):
        mention=pd.Series([row.quoteID,row.qids,row.speaker,mention_qid,row.mentions[j],row.country[j]])
        mentions_w_df=mentions_w_df.append(mention,ignore_index=True)
        if i %10000==0: print(i)
mentions_w_df.rename(columns={0:'quoteID',1:'qid',2:'mention_qid',3:'mentioned',4:'speaker',5:'country'},inplace=True)
print(mentions_w_df.head())
print(f'men_df shape {mentions_w_df.shape}')
```

```python
mentions_w_df.head(10)
```

```python
#change Danish Realm = Danemark (otherwhise can not have localisation)
mentions_w_df.loc[mentions_w_df.country=='Danish Realm','country']='Danemark'
```

```python
mentions_w_df['year'] = mentions_w_df['quoteID'].apply(lambda x: x[:4])
mentions_w_df['month'] = mentions_w_df['quoteID'].apply(lambda x: x[:7])#month maybe too much !!!! NEED TO BE ORDERED
mentions_w_df.head(10)
```

```python
year_m=mentions_w_df.groupby(['year'])
year_m.indices
```

```python
year_2=mentions_w_df.groupby(['year'])
exp=year_m2.get_group("2019")
exp2=exp.groupby(['country'])
len(exp2)
```

```python
years= ['2015','2016','2017','2018','2019','2020']

Coun_year=pd.DataFrame()

group_years=mentions_w_df.groupby(['year'])
for y in years:
    y1=year_m2.get_group(y)
    y2=y1.groupby(['country'])
    freq=y2.count().qid/len(mentions_w_df)#freq of number of countries mentioned that year
    lat=[]
    lon=[]
    for i,count in enumerate(y2.indices):
        geoloc=geolocator.geocode(count)[1]
        lat.append(geoloc[0])
        lon.append(geoloc[1])
        if i%40==0 : print(count)
    countries=pd.DataFrame({'country':freq.index,'freq':freq.values,'year':y,'lat':lat,'lon':lon})
    Coun_year=Coun_year.append(countries,ignore_index=True)
Coun_year.head()
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my_user_agent")
localisation=pd.DataFrame()
for i,row in countries.iterrows():
    
    
    locali=pd.Series([row.country,row.freq,geolocator.geocode(row.country)[1]])
    localisation=localisation.append(locali,ignore_index=True)
    if i%20==0 : print(row.country)
localisation.rename(columns={0:'country',1:'freq',2:'coordinates'},inplace=True)
print(localisation.head())
print(f'men_df shape {localisation.shape}')
```
