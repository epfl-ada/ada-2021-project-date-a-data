```python
#plot only USA and then check for global ...

quotes_mapped='quotebank_politicians_USA_mapped.json.bz2'
WIKI_DATA_FILTERED='filtered_politician_labeled_us_v3.json.bz2' #catalogue containing all US politician

df_full=pd.read_json(quotes_mapped, lines=True)
catalogue_df=pd.read_json(WIKI_DATA_FILTERED, lines=True)
catalogue_df.head(3)
```

```python
df_full.head(3)
```

```python
df=df_full[['quoteID','qids','mentions_qids','speaker']] #extract info of quotations taking only USA politicians mentions of USA politician in the US catalogue
#df=df.loc[:500,:]
men_df=pd.DataFrame()
for i,row in df.iterrows():
    if row.qids[0] in catalogue_df.qid.values:
        for mention_qid in row.mentions_qids:
            if mention_qid in catalogue_df.qid.values:
                mention=pd.Series([row.quoteID,row.qids[0],mention_qid,row.speaker])
                men_df=men_df.append(mention,ignore_index=True)
                if i %10000==0: print(i)
men_df.rename(columns={0:'quoteID',1:'qid',2:'mention_qid',3:'speaker'},inplace=True)
print(men_df.head())
print(f'men_df shape {men_df.shape}')
```

```python
#add state of reference (for world different )
#state_speaker = catalogue_df[catalogue_df['qid']==men_df['qid']]
#state_mention = catalogue_df[catalogue_df['qid']==men_df.mention_qid]
#men_df['speak_state']= catalogue.iloc['qid']

States=[catalogue_df.loc[catalogue_df.qid == qid, 'State'].item() for qid in men_df.qid]
States
```
